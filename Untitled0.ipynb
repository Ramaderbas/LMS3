```python
def contain_special_characters(password):
   char="!@#$%^&*()_+"
   return sum(char in password for char in char)

def contain_dijit(password):
   char="0123456789"
   return sum(char in password for char in char)

def run_the_system(dataset):
  while True:
     your_option= input("""Choose an number of option:
     [1]Create new user.
     [2]Log in.
     [3]Exist.""")
     if your_option == "1":
        create_new_user(dataset)
     elif your_option == "2":
        user_name = log_in(dataset)
        while user_name in dataset and dataset[user_name]["password"] == password:
          your_new_option=input("""Choose an number of new option:
        [1] Get student courses.
        [2] Add new user course.
        [3] Calculating the average.
        [4] Get_number of users.
        [5] Remove my account.
        [6] Log out.""")
          if your_new_option == "1":
            get_student_courses(user_name)
          elif your_new_option == "2":
            initial_grade=0
            add_new_user_course(user_name,initial_grade)
          elif your_new_option == "3":
            calculating_the_average(user_name)
          elif your_new_option == "4":
            get_number_of_users(dataset)
          elif your_new_option == "5":
            remove_my_account(user_name)
          elif your_new_option == "6":
            log_out(user_name)
          else:
            print(f"The option {your_new_option} is not of the possible options.")
          cont=input("If you want choose onther option write yes:")
          if cont.lower()=="yes":
             continue
          else:
             print("Exiting the system.")
             break
     elif your_option == "3":
        print("Exiting the system.")
        print(f"Bye {user_name}.")
        break

def log_in(dataset):
    user_name = input("Write your user name (must be lower latter): ")
    if user_name in dataset:
        global password
        password = input("Write your password: ")
        if dataset[user_name]["password"] == password:
            print("You are logged in")
            print(f"Hello {user_name}.I wish you success.")
            return user_name
        else:
            print("Wrong password, Try please.")
        return None
    else:
        print("user name not found")

def create_new_user(dataset):
    user_name = input("Write your user name: ").lower()
    if user_name not in dataset:
        password = input("Write your password: ")
        if len(password)>=16:
          if user_name in password:
            print("The password cannot contain the user name.")
          else:
            if contain_special_characters(password)  >=2 :
              if contain_dijit(password) >=2:
               dataset[user_name] = {"password": password, "grade courses": {}}
               print("The account has been created successfully")
               print("Hi my dear. Welcome in LMS")
              else:
               print("The password must contain at least 2 dijit.")
            else:
              print("The password must contain at least 2 special characters.")
        else:
          print("The password must be at least 16 characters.")


    else:
        print("user name already exists")

def get_student_courses(user_name):
    if user_name in dataset:
        courses = list(dataset[user_name]["grade courses"].keys())
        if len(courses) > 0:
           print("Your courses are:", courses)
        else:
          print("You are no have any courses yet.")

def add_new_user_course(user_name,initial_grade):
    course = input("Write the new course: ")
    initial_grade=int(input(f"Write the initail grade for {course} course: "))
    if user_name in dataset:
        courses = list(dataset[user_name]["grade courses"].keys())
        if course not in courses:
            dataset[user_name]["grade courses"][course]= initial_grade
            print(f"The new course {course} has been added with an initial grade of {initial_grade}.")
        else:
            print(f"The course {course} already exists.")


def calculating_the_average(user_name):
    if user_name in dataset:
        grades = list(dataset[user_name]["grade courses"].values())
        print("your grades in ordered courses are:", grades)
        if len(grades) > 0:
            average = sum(grades) / len(grades)
            print("Average grade:", average)
        else:
            print("No grades available.")

def get_number_of_users(dataset):
    num_of_user=len(dataset)
    print("the number of user in the system is: " , num_of_user)

def remove_my_account(user_name):
  if user_name in dataset:
    del dataset[user_name]
    print("your account has been removed")
  else:
    print("your account is not found")

def log_out(user_name):
  if user_name in dataset:
     print("you are logged out")
  else:
    print("the user account is not found")

dataset={}
run_the_system(dataset)
```
